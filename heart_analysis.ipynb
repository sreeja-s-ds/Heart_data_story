```python
import pandas as pd
```

```python
df=pd.read_csv('/content/heart_2020_cleaned data.csv')
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for i in df.select_dtypes(include=["object"]).columns:
  df[i]= le.fit_transform(df[i])
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.fillna(0,inplace=True)
```

```python
df[df.duplicated()]
```

```python
df['Race'].value_counts(normalize=True) * 100
# distribution of entries across different race groups?
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.countplot(data=df, y='Race', order=df['Race'].value_counts().index)
plt.title('Race Group Distribution (%)')
plt.show()
```

```python
df['GenHealth'].value_counts(normalize=True).iloc[['3', '2']] * 100
# percentage of people having general health marked as "Poor" or "Fair"
```

```python
sns.countplot(data=df, x='GenHealth', order=df['GenHealth'].value_counts().index)
plt.title('General Health Status')
plt.show()
```

```python
df['PhysicalHealth'].describe()
# physical health (number of physically unwell days) distributed in the population
```

```python
sns.histplot(df['PhysicalHealth'], bins=30, kde=True)
plt.title('Distribution of Physical Health (Unhealthy Days)')
plt.xlabel('Unhealthy Days in Past 30')
plt.show()
```

```python
df['Sex'].value_counts()
# counts of gender
```

```python
sns.countplot(data=df, x='Sex')
plt.title('Gender Distribution')
plt.show()
```

```python
df['AgeCategory'].value_counts().sort_index()
# average age category
```

```python
sns.countplot(data=df, y='AgeCategory', order=sorted(df['AgeCategory'].unique()))
plt.title('Age Category Distribution')
plt.show()
```

```python
(df['Stroke'] == 'Yes').mean() * 100
# percentage of people who had stroke
```

```python
df['Diabetic'].value_counts()
```

```python
sns.countplot(data=df, x='Diabetic')
plt.title('Diabetes Status Distribution')
plt.show()
```

```python
pd.crosstab(df['Smoking'], df['HeartDisease'], normalize='index') * 100
# Is there a relationship between Heart Disease and Smoking?
```

```python
sns.countplot(data=df, x='Smoking', hue='HeartDisease')
plt.title('Smoking vs Heart Disease')
plt.show()
```

```python
heart_rate_by_age = df[df['HeartDisease'] == 'Yes']['AgeCategory'].value_counts(normalize=True) * 100
heart_rate_by_age.sort_values(ascending=False)
# age group has the highest rate of heart disease
```

```python
sns.histplot(df['BMI'], kde=True, bins=30)
plt.title('BMI Distribution')
plt.show()
```
